# Organizing Data

```python
import skimage as ski
import skimage.io
import pylab as pl
import numpy as np

testImage = ski.io.imread('./tests/Example.tif')
pl.figure()
ski.io.imshow(testImage)
testImage = (testImage > 0)
center = np.array([116, 101])

[xdata, ydata] = np.where(testImage)

data = np.column_stack((xdata, ydata))
```

# Db clustering

```python
import sklearn
import sklearn.cluster
import scipy as sp

db = sklearn.cluster.DBSCAN(eps=1.5, min_samples=6).fit(data)
labels = db.labels_
# Make into a label image
labelImage = np.zeros(testImage.shape, dtype=np.int)

# We want labels to go from 1 upwards
labelImage[xdata, ydata] = labels + 2
imshow(labelImage)
```

# Hierarchical Ward Clustering

```python
# Get connectivity data
connectivity = sklearn.neighbors.kneighbors_graph(data, n_neighbors=8) 
ward = sklearn.cluster.Ward(n_clusters = 10, connectivity=connectivity).fit(data)
labels = ward.labels_

# Make into a label image
labelImage = np.zeros(testImage.shape, dtype=np.int)

# We want labels to go from 1 upwards
labelImage[xdata, ydata] = labels + 2
imshow(labelImage)
```

#Affinity Clustering

```python
af = sklearn.cluster.AffinityPropagation().fit(data)
labels = af.labels_
# Make into a label image
labelImage = np.zeros(testImage.shape, dtype=np.int)

# We want labels to go from 1 upwards
labelImage[xdata, ydata] = labels + 2
imshow(labelImage)
```
