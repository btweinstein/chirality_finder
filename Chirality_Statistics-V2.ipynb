## Getting error bars

```python
import scipy as sp
import scipy.stats
```

Ok, I am having some trouble dealing with the different types of columns. Let's just extract the ones we need from all_Data and go from there.

```python
def apply_sem_to_numeric(x):
     return x._get_numeric_data().apply(axis=0, func=sp.stats.sem)
```

```python
stder_data = group_T.apply(apply_sem_to_numeric)
stder_data = stder_data.add_suffix('_stderr')
display(stder_data)
```

Beautiful. The key was to only apply functions to the numeric data, or else everything get screwed up. Let's make sure things look reasonable.

```python
stder_data.head(5)
```

We are only interested in rotated, really, and that looks reasonable so we are all set. Let's now replot with error bars.

```python
mean_T_Data = meanData.reset_index()
stder_T_data = stder_data.reset_index()

# We now want to join the data by radius interval and temperature. 

combinedData = pd.merge(mean_T_Data, stder_T_data, on=['T', 'r_mm'])
display(combinedData)
```
