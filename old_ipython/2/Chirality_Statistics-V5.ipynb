# Chirality vs. Temperature

Last time, we used pixel coordinates for r when averaging. This is not a good idea as each image had a different zoom. How do we make the jump from pixel coordinates to actual physical distance? We probably need to bin the actual distance, unfortunately. Yeah, let's use the same strategy we did in my chirality data analysis program. 

The question is how many bins should we use. Let's just plot and find out what looks reasonable.

```python
minPhysicalR = allData['r_mm'].min() - 0.001
maxPhysicalR = allData['r_mm'].max() + 0.001
numBins = 100

print 'Min radius:', minPhysicalR , 'mm'
print 'Max radius:', maxPhysicalR , 'mm'

bins, binsize = np.linspace(minPhysicalR, maxPhysicalR, numBins, retstep=True)

print 'Binsize:' , binsize

group_T = allData.groupby([pd.cut(allData['r_mm'], bins), 'T'])
```

```python
meanData = group_T.mean()
meanData = meanData.add_suffix('_mean')
```

```python
display(meanData.head(2))
```

Ignoring error bars, let's see what happens here. I am going to use $r d\theta$ as my assay for chirality here as I think it will be more fair physically, i.e. far more independent on where I choose to measure the zero of my coordinate system.

```python
mean_T_Data = meanData.reset_index()
meanDataGroups = mean_T_Data.groupby('T')
for name, df in meanDataGroups:
    df.plot(x='r_mm_mean', y='rdtheta_mm_mean', style= '+', label='T=' + str(name))
plt.xlabel('Radius (mm)')
plt.ylabel('Average $r d\\theta$ (mm)')
plt.title('Average $r d\\theta$ vs. Radius')
plt.legend(loc='best')
```

Wow! There does seem to be noticable more chirality in the high temperature case! The question is if this is accurate or not; we will have to use error bars! We will use the standard error of the mean.

## Getting error bars

```python
import scipy as sp
import scipy.stats
```

Ok, I am having some trouble dealing with the different types of columns. Let's just extract the ones we need from all_Data and go from there.

```python
def apply_sem_to_numeric(x):
     return x._get_numeric_data().apply(axis=0, func=sp.stats.sem)
```

```python
stder_data = group_T.apply(apply_sem_to_numeric)
stder_data = stder_data.add_suffix('_stderr')
```

Beautiful. The key was to only apply functions to the numeric data, or else everything get screwed up. Let's make sure things look reasonable.

```python
stder_data.head(2)
```

We are only interested in rotated, really, and that looks reasonable so we are all set. Let's now replot with error bars.

```python
mean_T_Data = meanData.reset_index()
stder_T_data = stder_data.reset_index()

# We now want to join the data by radius interval and temperature. 

combinedData = pd.merge(mean_T_Data, stder_T_data, on=['T', 'r_mm'])
```

Let's make a nicer plot.

```python
combinedDataGroups = combinedData.groupby('T')

counter = 0
colors = ['r', 'g', 'b']

for name, df in combinedDataGroups:
    x = df['r_mm_mean'].values
    y = df['rdtheta_mm_mean'].values
    xerr = binsize
    yerr = df['rdtheta_mm_stderr'].values
    
    plt.plot(x , y, '+-', label=name, color=colors[counter])
    plt.fill_between(x, y - yerr, y + yerr, alpha=0.3, antialiased=True, color=colors[counter])
    # Only fill in regions where it hasn't been filled already or things will look weird
    #plt.fill_betweenx(y, x - xerr, x + xerr, alpha=0.3, antialiased=True, color=colors[counter])
    counter += 1
plt.xlabel('Radius (mm)')
plt.ylabel('Average d$\\theta$')
plt.title('Average d$\\theta$ vs. Radius')
#plt.ylim([-0.02, 0.10])
plt.legend(loc='best')
```

# Duplicating Korolev's Plots

This really should have been from the get go. I don't remember why I didn't do this. That's ok. I probably forgot about the plots.

* We need to calculate $\frac{r}{r_i}$, so we need r_i for each sector.

```python
path_label_groups = allData.groupby(['path', 'label'])
```

We need to get the minimum radius. How can we do this?

```python
minRadii = path_label_groups.agg({'r_mm' : min})
minRadii = minRadii.add_suffix('_min')
display(minRadii.head(5))
```

Great. We got it! Let's now combine the two datasets.

```python
mr_reset = minRadii.reset_index()
ad_with_rmin = pd.merge(allData, mr_reset, on=['path', 'label'])
print ad_with_rmin
```

Let's now make Korelov's first plot. We'll have to do some tweaks, of course, i.e. playing with the data.

```python
ad_with_rmin['log_r_div_ri'] = np.log(ad_with_rmin['r_mm']/ad_with_rmin['r_mm_min'])
```

```python
g_rmin = ad_with_rmin.groupby('T')
meanData = g_rmin.mean()
meanData = meanData.add_suffix('_mean')
display(meanData)
```

```python
groups = ad_with_rmin.groupby('T')
for name, group in groups:
    newgroup = group.groupby(['path', 'label'])
    plt.figure()
    for n, g in newgroup:        
        g = g.sort('r_mm')
        x = np.log(g['r_mm']/g['r_mm_min'])
        y = g['rotated_righthanded']
        plt.plot(x, y, '-')
    plt.title(name)
    plt.xlabel('$\\ln(r/r_i)$')
    plt.ylabel('d$\\theta$')
```
