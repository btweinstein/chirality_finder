We simply need to figure out the radius of every image in this table.

```python
from chirality_image_analysis import image_analysis as chi
from chirality_image_analysis import utility as ut
import skimage as ski
import skimage.io
```

```python
imageData['colony_radius_mm'] = np.NaN
imageData['colony_radius_mm_stderr'] = np.NaN
numImages = len(imageData)
```

```python
growthOutputFolder = '/home/bryan/Documents/Research_Data/Nelson/Chirality_Initial_Imaging/Analysis/Growth/Automated_Output'
editedFolder = '/home/bryan/Documents/Research_Data/Nelson/Chirality_Initial_Imaging/Analysis/Growth/Edited_Output'
```

```python
ski.io.use_plugin('tifffile')
```

```python
count = 0
for i in range(numImages):
    
    currentRow = imageData.iloc[i]
    img = ski.io.imread(currentRow['path'])
    imageName = os.path.basename(currentRow['path'])
    print imageName
    
    scale = ut.getScale(currentRow['path'])
    fluor1 = img[:, :, 0]
    fluor2 = img[:, :, 1]
    brightfield = img[:, :, 2]
    
    output = chi.findBrightfieldCircle(brightfield, returnOverlay=True)
    
    radius = output[1]
    cov = output[2]
    overlayImage = output[3]
    
    imageData['colony_radius_mm'].iloc[i] = radius*scale
    radiusError = np.sqrt(cov[2, 2]) * scale
    imageData['colony_radius_mm_stderr'].iloc[i] = radiusError
    
    colorBrightfield = ski.color.gray2rgb(brightfield)
    
    combinedImage = np.array([overlayImage, colorBrightfield])
    # Create the image stack that you will save
    ski.io.imsave(growthOutputFolder + '/' + imageName, combinedImage, plugin='tifffile')
    count += 1
```

# Importing the Data

```python
imageData = pd.read_csv(growthOutputFolder + '/automated_Output.csv')
imageData = imageData.drop('Unnamed: 0', 1)
display(imageData)
```

```python
import glob
import scipy as sp
import scipy.stats
```

```python
files = glob.glob(editedFolder + '/*.txt')
editedData = pd.DataFrame()
for f in files:
    data = pd.read_csv(f, sep='\t')
    data['short_name'] = os.path.basename(f)
    data['short_name'] = [f.split('.')[0] for f in data['short_name']]
    data = data.drop([' ', 'Angle'], 1)

    editedData = editedData.append(data)
editedData = editedData.reset_index(drop=True)

display(editedData.head(5))
```

Good, that is working. Let us now combine this with the old imageData table.

```python
imageData['short_name'] = imageData.path.apply(os.path.basename)
imageData['short_name'] = [f.split('.')[0] for f in imageData['short_name']]
imageData.head(2)
```

```python
combinedData = pd.merge(imageData, editedData, on='short_name', how='outer')
display(combinedData.head(5))
```

Beautiful. Let us now get an appropriate scale for each image.

```python
combinedData['scale'] = combinedData['path'].apply(ut.getScale)
display(combinedData.head(1))
```

```python
de = ~ combinedData.Length.isnull()
combinedData['colony_radius_mm'][de] = combinedData['Length'][de]*combinedData['scale'][de]
combinedData.head(3)
```

```python
growthData = combinedData.groupby(['path', 'date', 'T', 'mL', 'name']).agg([mean, sp.stats.sem])
growthData = growthData.reset_index()
display(growthData.head(3))
```

Cool. Let's organize them by temperature.

```python
growthData = growthData.sort('date')
groups = growthData.groupby(['T', 'mL', 'name'])
numGroups = len(groups)
count = 0
for name, group in groups:
    if name[0] == 21:
        group.plot(x='date', y=('colony_radius_mm', 'mean'), style='+-', color='r', figsize=(12, 6))
    if name[0] == 30:
        group.plot(x='date', y=('colony_radius_mm', 'mean'), style='+-', color='g', figsize=(12, 6))
    if name[0] == 37:
        group.plot(x='date', y=('colony_radius_mm', 'mean'), style='+-', color='b', figsize=(12, 6))
    count += 1
```

```python
growthData = growthData.sort('date')
groups = growthData.groupby(['T'])
numGroups = len(groups)
for temp, tempGroup in groups:
    plt.figure()
    tempGroup.sort('mL')
    otherGroups = tempGroup.groupby(['mL', 'name'])
    count = 0
    numColors = len(otherGroups)
    for name, g in otherGroups:
        g.plot(x='date', y=('colony_radius_mm', 'mean'), style='+-', figsize=(12,6), label=name, color=plt.cm.jet(1.0*count/numColors))
        count += 1
    plt.legend(loc='best')
    plt.title('T= ' + str(temp))
    plt.ylabel('Radius (mm)')
```
