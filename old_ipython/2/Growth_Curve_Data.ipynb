Cool. Let's organize them by temperature.

```python
imageData = imageData.sort('date')
groups = imageData.groupby(['T', 'mL', 'name'])
numGroups = len(groups)
count = 0
for name, group in groups:
    if name[0] == 21:
        group.plot(x='date', y='colony_radius_mm', style='+-', color='r', figsize=(12, 6))
    if name[0] == 30:
        group.plot(x='date', y='colony_radius_mm', style='+-', color='g', figsize=(12, 6))
    if name[0] == 37:
        group.plot(x='date', y='colony_radius_mm', style='+-', color='b', figsize=(12, 6))
    count += 1
```

Well, as expected, 37 grew the fastest, followed by 30, followed by 21. Cool! Let's see within each of these plots if volume made a big difference.

```python
imageData = imageData.sort('date')
groups = imageData.groupby(['T'])
numGroups = len(groups)
for temp, tempGroup in groups:
    plt.figure()
    tempGroup.sort('mL')
    otherGroups = tempGroup.groupby(['mL', 'name'])
    count = 0
    numColors = len(otherGroups)
    for name, g in otherGroups:
        g.plot(x='date', y='colony_radius_mm', style='+-', figsize=(12,6), label=name, color=plt.cm.jet(1.0*count/numColors))
        count += 1
    plt.legend(loc='best')
    plt.title('T= ' + str(temp))
    plt.ylabel('Radius (mm)')
```
