We now have to apply our chirality finder program to a large lump of data and organize the results. What should our strategy be here?

```python
import chirality_data_analysis as chd
from chirality_data_analysis import example_data as ex
import pandas as pd
```

```python
imageData = ex.imageData
display(imageData.head(10))
```

```python
groups = imageData.groupby(['T','mL'])
for name, group in groups:
    print name
    display(group)
```

Ok, that worked well. We now have the different groups organize by temperature and volume. We want to analyze each of these groups, likely separately, and store the results in a big table. Let's do an example group.

```python
import matplotlib
import matplotlib.pyplot as plt
from chirality_data_analysis.example_data import *
from chirality_image_analysis import image_analysis as chi
from chirality_image_analysis.utility import *
from chirality_data_analysis import data_analysis as chd
import skimage as ski
import skimage.color
import skimage.io

analysisFolder = '/home/bryan/Documents/Research_Data/Nelson/Chirality_Initial_Imaging/Analysis/Chirality'
chiralityDataName = 'chiralityData.png'
chiralityPlotName = 'chiralityPlot.png'
sectorPlotName = 'sectorPlot.png'
overlayPlotName = 'overlayPlot.png'
```

```python
print len(groups)
```

So there are 9 groups to work with here. Let us work with each separately, as we will likely have to tweak the options for each. We are only interested in the pictures that have the maximum date for the chirality. Right? Well, we can start there, we may need the others for growth data in the future.

```python
largestDates = imageData.sort('date').groupby(['T','mL']).last()
largestDates = largestDates.reset_index()
display(largestDates)
```

Let's make a convenience function to create the folder where the output will lie.

```python
def getFolderName(row):
    return analysisFolder + '/' + str(row['T']) + '_' + str(row['mL']) + '_' + str(row['name'])

getFolderName(largestDates.iloc[0])
```

Ok, that works well. The idea here is that we do each example individually.

## Index 0

```python
import os

row = largestDates.iloc[0]
rowDirectory = getFolderName(row)

print rowDirectory

# Check if folder exists
if not os.path.exists(rowDirectory):
    print 'Made new directory!'
    os.makedirs(rowDirectory)
else: print 'Directory already exists!'
```

```python
img = ski.io.imread(row['path'])
fluor1 = img[:, :, 0]
fluor2 = img[:, :, 1]
brightfield = img[:, :, 2]

filteredLabels, center, finalRadius = chi.findSectors(fluor1, brightfield, showPictures=False, \
                    homelandCutFactor=0.4, edgeCutFactor=0.9, threshold_factor=1.0)
```

```python
# Now that we have the labels, we need to get the position data of each label
# and filter them all down to one pixel
chiralityData = chd.getChiralityData(filteredLabels, center)

chiralityData.to_csv(rowDirectory + '/' + chiralityDataName)

chiralityPlot = chd.makeChiralityPlot(chiralityData)
chiralityPlot.savefig(rowDirectory + '/' + chiralityPlotName)

chd.visualizeSectors(chiralityData)
sectorPlot.savefig(rowDirectory + '/' + sectorPlotName)

# Overlay the sectors on the image now
chd.visualizeSectors(chiralityData, overImage=True)
# Overlay the center
plt.plot(center[1], center[0], 'o')
# Plot the image
imagePlot = ski.io.imshow(img, interpolation='None', cmap=plt.cm.YlGnBu)
plt.gcf().savefig(rowDirectory + '/' + overlayPlotName)

plt.close('all')

# Now display the images since pylab inline kills everything
display(Image(filename=rowDirectory + '/' + chiralityPlotName, retina=True))
display(Image(filename=rowDirectory + '/' + sectorPlotName, retina=True))
display(Image(filename=rowDirectory + '/' + overlayPlotName, retina=True))
```
