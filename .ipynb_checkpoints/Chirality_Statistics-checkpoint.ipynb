# Chirality vs. Temperature

This is the statistic that we are most interested in. It should be relatively easy. Why don't we plot average $d\theta$ vs. $r$ at each temperature. That is probably the easiest way to quantify this. We will have to put on error bars at some point too.

```python
group_T = allData.groupby(['T', 'radius_interval'])
```

```python
meanData = group_T.mean()
meanData = meanData.add_suffix('_mean')
print meanData
```

```python
display(meanData.head(10))
```

Ignoring error bars, let's see what happens here.

```python
mean_T_Data = meanData.reset_index()
meanDataGroups = mean_T_Data.groupby('T')
for name, df in meanDataGroups:
    df.plot(x='r_mean', y='rotated_mean', style= '+', label='T=' + str(name))
plt.legend(loc='best')
plt.xlabel('r')
plt.ylabel('Average d$\\theta$')
plt.title('Average d$\\theta$ vs. r')
plt.ylim([-0.01, 0.07])
```

Wow! There does seem to be noticable more chirality in the high temperature case! The question is if this is accurate or not; we will have to use error bars! We will use the standard error of the mean.

TODO: We should probably throw out points that do not have enough data, but we can deal with that later.

## Getting error bars

```python
import scipy as sp
import scipy.stats
```

Ok, I am having some trouble dealing with the different types of columns. Let's just extract the ones we need from all_Data and go from there.

```python
def apply_sem_to_numeric(x):
     return x._get_numeric_data().apply(axis=0, func=sp.stats.sem)
```

```python
stder_data = group_T.apply(apply_sem_to_numeric)
stder_data = stder_data.add_suffix('_stderr')
display(stder_data)
```

Beautiful. The key was to only apply functions to the numeric data, or else everything get screwed up. Let's make sure things look reasonable.

```python
stder_data.head(5)
```

We are only interested in rotated, really, and that looks reasonable so we are all set. Let's now replot with error bars.

```python
mean_T_Data = meanData.reset_index()
stder_T_data = stder_data.reset_index()

# We now want to join the data by radius interval and temperature. 

combinedData = pd.merge(mean_T_Data, stder_T_data, on=['T', 'radius_interval'])
display(combinedData)
```

We now have all of the data in one dataframe. We are now ready to plot!

```python
combinedDataGroups = combinedData.groupby('T')
for name, df in combinedDataGroups:
    plt.errorbar(df['r_mean'], df['rotated_mean'], yerr=df['rotated_stderr'], fmt='+', label=name)
plt.legend(loc='best')
plt.xlabel('r')
plt.ylabel('Average d$\\theta$')
plt.title('Average d$\\theta$ vs. r')
plt.ylim([-0.01, 0.1])
```

This looks like gobbly-gook. Let's make a nicer plot. 

```python
combinedDataGroups = combinedData.groupby('T')

numColors = len(combinedDataGroups)
counter = 0
colormapToUse = plt.cm.jet

for name, df in combinedDataGroups:
    x = df['r_mean']
    y = df['rotated_mean']
    yerr = df['rotated_stderr']
    plt.plot(x , y, '+-', label=name, color=colormapToUse(1.*counter/numColors))
    plt.fill_between(x, y - yerr, y + yerr, interpolate=True, alpha=0.3, antialiased=True, color=colormapToUse(1.*counter/numColors))
    counter += 1
plt.legend(loc='best')
plt.xlabel('r')
plt.ylabel('Average d$\\theta$')
plt.title('Average d$\\theta$ vs. r')
plt.ylim([-0.01, 0.08])
```

That was irritating. Using colormaps does not always work how I want. Let's just specify the colors ourselves.

```python
combinedDataGroups = combinedData.groupby('T')

counter = 0
colors = ['r', 'g', 'b']

for name, df in combinedDataGroups:
    x = df['r_mean']
    y = df['rotated_mean']
    yerr = df['rotated_stderr']
    plt.plot(x , y, '+-', label=name, color=colors[counter])
    plt.fill_between(x, y - yerr, y + yerr, interpolate=True, alpha=0.3, antialiased=True, color=colors[counter])
    counter += 1
plt.legend(loc='best')
plt.xlabel('r')
plt.ylabel('Average d$\\theta$')
plt.title('Average d$\\theta$ vs. r')
plt.ylim([-0.01, 0.08])
```

There is a LOT of error. It is probably more instructive to just plot the spread of points so that we can see the trend instead of using error bars.

# Looking at distribution of Error

Why don't we look at the error probability distribution vs. radius interval. We can make some sort of histogram. We ideally want to see how the histogram changes with radius interval. 

```python
count = 0

rows = 10
columns=5
plotNum = 1

plt.figure(figsize=(18, 18))
for name, group in group_T:
    if mod(count, 20) == 0:
        plt.subplot(rows, columns, plotNum, title=name)
        plt.hist(group['rotated'])
        plt.grid()
        plt.xlim([-0.3, 0.3])
        plotNum += 1
    count += 1
    
plt.subplots_adjust(wspace=0.5, hspace=0.5)
```

The issue with the 37 data is that there is much less of it than everything else. Let us show this. In fact, let us plot the number of sectors in a given radius per temperature.
