# Approximating Growth Velocity with PyMC

## Converting time to delta T

Let us convert the time into seconds.

```python
import time
```

```python
growthData['time_s'] = growthData['date'].apply(lambda x: time.mktime(x.timetuple()))
```

```python
plate_groups = growthData.groupby(['T', 'mL', 'name'])
timeDelta = plate_groups['time_s'].transform(lambda x: x - x.min())
timeDelta.name = 'timeDelta'
print timeDelta.head(15)

combinedData = growthData.copy()
combinedData['timeDelta'] = timeDelta
```

```python
combinedData.head(15)
```

Good. Finally! That works. lol. We now have the deltaT in seconds.

## Testing PYMC

I created, by hand, the linear regression. Let's see if we can do the same thing with PyMC.

```python
import pymc
```

Let us first attempt to do a simple linear regression of our data. Let's get a dataset.

```python
plate_groups = combinedData.groupby(['T', 'mL', 'name'])
```

```python
for name, current_group in plate_groups:
    break
```

```python
display(current_group)
```

Excellent. We now have our current group. Let us first make a list of $R$ variables. Actually, to make our life easy, let's just do a bayesian regression to fit $v_{||}$ as I do not belive the error we got currently.

```python
# Priors
sigma = pymc.HalfNormal('sigma', tau=1./(0.5)**2, value=0.1)
ro = pymc.HalfNormal('ro', tau=1./(1.5)**2, value=1.0)
vpar = pymc.Uniform('vpar', lower=0, upper=1, value=10**-6)
t = current_group['timeDelta'].values

@pymc.deterministic
def modeled_R(ro=ro, vpar=vpar, t=t):
    return ro + vpar*t

R = pymc.Normal('R', mu = modeled_R, tau=1.0/sigma**2., value=current_group['colony_radius_mm'], observed=True)
```

```python
M = pymc.MCMC([sigma, ro, vpar, t, R])
M.sample(200000, burn=10000, thin=50)
```

```python
pymc.Matplot.plot(M)
```

```python
maxValues = pymc.MAP([sigma, ro, vpar, t, R])
maxValues.fit(iterlim=100000)
```
